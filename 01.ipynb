Average cost per company

```python
avg_cost_per_company = product_data.groupby('company')['cost'].mean().reset_index() # We use resetindex to make it into dataframe
avg_cost_per_company
```

```python
type(avg_cost_per_company)
```

Create a barplot where in x-axis we have company and in y-axis we have average cost per company

```python
plt.figure(figsize=(18, 12))
plt.bar(avg_cost_per_company['company'], avg_cost_per_company['cost'], color= 'red')
plt.title('Average product cost per company', fontsize = 18)
plt.xlabel('company', fontsize = 16)
plt.ylabel('Average Cost', fontsize = 16)
plt.xticks(rotation=45)   # its rotates the x-axis text
plt.yticks(rotation=45)   # its rotates the y-axis text
plt.grid(axis='y', linestyle='--', linewidth=0.5, color='blue')
plt.savefig('average_product_cost.png')  # To save the figure

plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot(product_data['cost'])

plt.title('Box plot of cost of the product', fontsize=16)
plt.ylabel('Cost of producut')

plt.show()
```

Visualize the customers spending pattern over the time on a month basis using line plot.

```python
purchases_data.head()
```

```python
purchases_data_by_month = purchases_data.copy()
purchases_data_by_month.head()
```

```python
purchases_data_by_month['purch_month'] = pd.to_datetime(purchases_data_by_month['purch_date']).dt.month
purchases_data_by_month
```

```python
spending_pattern = purchases_data_by_month.groupby('purch_month')['amount'].sum().reset_index()
spending_pattern
```

```python
plt.figure(figsize=(15, 10))
plt.plot(spending_pattern['purch_month'], spending_pattern['amount'], color='blue', marker='o', )
plt.title("Customers spending patterns over time (Monthly)")
plt.xlabel("Month")
plt.ylabel("Total Amount Spend")
plt.grid('both', linestyle = "--", linewidth=0.6)
plt.show()
```

# Seaborn


> Simply built on top of matplotlib only



```python
tips = sns.load_dataset('tips')
tips
```

Total tips per day

```python
sns.barplot(x='day', y='tip', data=tips, estimator=sum)
plt.title('Total tips by day')
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.show()
```

KDE plots: Visualize the probability density function of any given variable

```python
sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True)
plt.title('KDE plot of total bill')
plt.xlabel('Total bill')
plt.show()
```

Box plot

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.title('Distributions pf tip per day')
plt.xlabel("Day")
plt.ylabel('Tips')
plt.show()
```

**Tast**



> Explore about the violinplot (detects the outliers just like boxplot)



```python
sns.violinplot(x='day', y='tip', data=tips)
plt.title('Distributions pf tip per day')
plt.xlabel("Day")
plt.ylabel('Tips')
plt.show()
```
